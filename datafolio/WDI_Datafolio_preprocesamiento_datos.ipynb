````
def evolution_i(n):
    df1 = df[df['Country Name'] == n]
    fig = px.area(df1, x="Year", y="(% of GDP)", facet_col="Indicator Name_x", facet_col_wrap=3, color='Indicator Name_x', title="Natural resources contribution to GDP")

    fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig.for_each_annotation(lambda a: a.update(text=a.text.split("(")[0]))
    fig.update_layout(showlegend=False)
    fig.show()
    return

country=widgets.Dropdown(
            options=sorted(df['Country Name'].unique().tolist()),
            description='Country Name:',
            disabled=False,
        )

interact(evolution_i,n=country)

#evolution_i('Afghanistan')
````
